1. <h4>Number of Engines</h4>
<h5>We will calculate the skewness of Number of Engines. If the variable is left or right skewed, we will replace null values with the median value. If the variable has a normal distribution we will replace null values with the mean.</h5>

```python
#calculating skewness of Number of Engines
skewness = df['Number.of.Engines'].skew()
print(f"Skewness: {skewness}")
```

verdict: Right skewed. replace null values with the median

```python
#replacing null values in Number of engines with the median
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(df['Number.of.Engines'].median())
#checking whether the null values in Number of engines have been replaced by the median
df.isna().sum()
```

<h3>Checking for Outliers</h3>

```python
df.columns
```

```python
sns.boxplot(df);
```

We have outliers in the following variables:

    1. Total Fatal Injuries
    2. Total Serious Injuries
    3. Total Minor Injuries
    4. Total Uninjured

```python

# List of columns to remove outliers from
columns = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']

# Creating a copy of the original DataFrame to preserve it
df_no_outliers = df.copy()

# Loop over the columns and remove outliers for each one
for col in columns:
    # Calculating the 0.5th percentile (lower bound) and 99.5th percentile (upper bound)
    lower_bound = df_no_outliers[col].quantile(0.005)  # 0.5th percentile
    upper_bound = df_no_outliers[col].quantile(0.995)  # 99.5th percentile
    
    # Filtering the data to remove outliers
    df_no_outliers = df_no_outliers[(df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)]
    
    # Plot a boxplot for the current column
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df_no_outliers[col])
    plt.title(f'Boxplot of {col} After Removing Outliers')
    plt.show()
```

<h5>Upon checking the distribution of the four columns, we have removed the existing outliers. </h5>

<h4>Data Transformation and Feature Engineering</h4>

    We'll create a "Total Injuries" column based on the weight of each injury type (fatal > serious > minor).<br>
    
    Based on my Business question at hand it is very important to create a weighted average of the Fatal, Serious and Minor Injuries<br>
    
    fatal > serious > minor This is because we are majorly looking for the safest airplane.<br>
    
    Assigning Fatal a higher weight allows us to get an unbiased outcome 


```python
# Calculate total injuries with weighted scores
cleandf['Total_Injuries'] = cleandf['Total.Fatal.Injuries'] * 5 + cleandf['Total.Serious.Injuries'] * 3 + cleandf['Total.Minor.Injuries'] * 1
```

```python
cleandf.columns
```

```python
#As seen above we now have a new column named Total_injuries
```

```python

# Converting categorical columns to category type
cleandf['Make'] = cleandf['Make'].astype('category')
cleandf['Engine.Type'] = cleandf['Engine.Type'].astype('category')
cleandf['Weather.Condition'] = cleandf['Weather.Condition'].astype('category')
cleandf['Injury.Severity'] = cleandf['Injury.Severity'].astype('category')

```

<h2>Multivariate Visualizations</h2>


Multivariate visualizations help explore the relationships between three or more variables.


```python
# Calculate correlation matrix
correlation_matrix = cleandf[['Total_Injuries', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 
                          'Total.Minor.Injuries', 'Number.of.Engines']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
